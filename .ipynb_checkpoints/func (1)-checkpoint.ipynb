```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def lscm(x, y, theta):

  x_next = np.sin(np.pi * (4 * theta * x * (1 - x) + (1 - theta) * np.sin(np.pi * y)))
  y_next = np.sin(np.pi * (4 * theta * y * (1 - y) + (1 - theta) * np.sin(np.pi * (x_next))))
  return x_next, y_next

def generate_lscm_secret_key():

  # Generate random initial states.
  x_0 = 0.8
  y_0 = 0.5
  theta = 0.99
  # Generate a random growth rate.
  r = 1.19 # initial r values

  # Generate random coefficients for the control parameters.
  p_1 = np.random.rand()      # for round 1
  p_2 = np.random.rand()      # for round 2

  # Return the secret key elements.
  return x_0, y_0, r, p_1, p_2, theta
```

```python
img = cv2.imread('heart.png', cv2.IMREAD_GRAYSCALE)
```

```python
secret_key = generate_lscm_secret_key()

# Initialize the LSCM chaotic map with the secret key.
x = secret_key[0]
y = secret_key[1]
r = secret_key[2]
p1 = secret_key[3]
p2 = secret_key[4]
theta = secret_key[5]
# theta = r_cp = (r*p1)%1  # theta = r_cp
# Generate a sequence of LSCM chaotic map values.
chaos = np.zeros(img.shape)
x_sequence = []
y_sequence = []
# for i in range(50):
#   for j in range(50):
for i in range(len(img)):
  for j in range(len(img[0])):
    x_next, y_next = lscm(x, y, theta) # Round 1 only as of now  # rcp is the control paramater or theta
    x, y = x_next, y_next
    chaos[i][j] = x
#    chaos_y[i][j] = y0  # chaos matrix for y which can be combined to create position matrix PM instead of using cols
    x_sequence.append(x_next)
    y_sequence.append(y_next)
```

```python
# Plot the LSCM chaotic map sequence - 50*50 samples
plt.plot(x_sequence, y_sequence, 'p')
plt.xlabel('x')
plt.ylabel('y')
plt.title('LSCM Chaotic Map')
plt.show()
```

```python
# Arnold cat map
def arnold_cat_map(image, iterations):
  height, width = image.shape

  # Pad the image to a square if necessary.
  if height != width:
    image = np.pad(image, [[0, 0], [0, width - height]], 'constant', constant_values=0)

  # Apply the map for the given number of iterations.
  for i in range(iterations):
    new_image = np.zeros_like(image)
    for y in range(height):
      for x in range(width):
        new_x = (2 * x + y) % width
        new_y = (x + y) % height
        new_image[new_y, new_x] = image[y, x]

    image = new_image

  return image
```

```python
transformed_image = arnold_cat_map(img, 10)

print(transformed_image.shape)
```

```python
trn_img = transformed_image.astype(np.uint8)
```

```python
cv2.imshow('transformed img', trn_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Permutaion
def generate_index_matrix(chaotic_matrix):
    index_matrix = np.argsort(chaotic_matrix, axis=1)
    return index_matrix

# Step 4: Form position matrix 'PM' using the index matrix 'I' as the row position
def generate_position_matrix(ind_m, size):
    position_matrix = [[None]*size[0] for i in range(size[1])]
    
    for i in range(size[0]):
        for j in range(size[1]):
            position_matrix[i][j] = [ind_m[i][j], j]
    return position_matrix

# Step 6: Select pixels from the plain image according to positions in the index matrix
def permute_image(img, position_matrix, chaotic, size):
    permuted_image = np.zeros(img.shape)
    for i in range(size[0]):
        # From each row of chaotic matrix take vector and sort, then put the selected pixels from plain image using PM(position_matrix values) into each position in that vector.
        e = np.argsort(chaotic[i])
        for j in range(size[1]):
            x, y = position_matrix[i][j]
            permuted_image[i][e[j]] =  img[x][y]
    return permuted_image

# Main function to perform permutation
def perform_image_permutation(plain_image):
    image_size = plain_image.shape
    #chaotic_matrix = chaos
    index_matrix = generate_index_matrix(chaos)
    position_matrix = generate_position_matrix(index_matrix, image_size)

    permuted_image = permute_image(plain_image, position_matrix, chaos, image_size)

    return permuted_image
```

```python
permuted_image = perform_image_permutation(trn_img)  
```

```python
perm = permuted_image.astype(np.uint8)
```

```python
cv2.imshow('permuted img', perm)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Diffusion
def diffusion(perm, chaotic, img, F):
    height, width = perm.shape
    encrypted_image = np.zeros_like(perm)
    key = [np.random.randint(0,255), np.random.randint(0,255)]
    bit=2**32
    for i in range(height):
        for j in range(width):
            if i == 0 and j==0:
                # int() is provided to take the intger of that chaotic value
                E_i = ((perm[0] + key[-1] + key[-2] + int((chaotic[i, j] * bit))) % F)[0]  # Because on adding we get in array format of size one, so we index it as 0
            elif i ==0 and j == 1:
                E_i = ((perm[1] + encrypted_image[i, j-1] + key[-1] + int((chaotic[i, j] * bit))) % F)[0]
            else:
                if j==0:# do this with j==0 or j==1
                    E_i = ((perm[i] + encrypted_image[i-1, width-1] + encrypted_image[i-1, width-2] + int((chaotic[i, j] * bit))) % F)[0]
                elif j==1:
                    E_i = ((perm[i] + encrypted_image[i, j-1] + encrypted_image[i-1, width-1] + int((chaotic[i, j] * bit))) % F)[0]
                else:
                    E_i = ((perm[i] + encrypted_image[i, j-1] + encrypted_image[i, j-2] + int((chaotic[i, j] * bit))) % F)[0]
            
            encrypted_image[i, j] = E_i
    return encrypted_image
```

```python
F = 256
encrypted_image = diffusion(permuted_image, chaos, img, F)
```

```python
encry = encrypted_image.astype(np.uint8)
```

```python
cv2.imshow('encrypted img', encry)
cv2.waitKey()
cv2.destroyAllWindows()
```
